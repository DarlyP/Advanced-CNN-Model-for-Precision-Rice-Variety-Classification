## ***Import Libraries***

```python
import numpy as np
from tensorflow.keras.preprocessing.image import img_to_array
from PIL import Image
from tensorflow.keras.models import load_model
```

```python
# Menentukan Path
best_model_path = '/content/drive/MyDrive/GC7/cnn_model.keras'
```

```python
# Impor Model
model = load_model(best_model_path)
```

## ***Prediction***

---

```python
# Mendefinisikan dimensi gambar
img_height, img_width = 220, 220

# Fungsi untuk mempreproses gambar
def preprocess_image(img_path):
    img = Image.open(img_path)
    img = img.resize((img_height, img_width))
    img_array = img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)  # Menambahkan dimensi batch
    img_array /= 255.0  # Menormalkan gambar ke rentang [0, 1]
    return img_array
```

```python
# Memuat gambar dari Google Drive
img_path = '/content/drive/MyDrive/GC7/image_1.jpg'

# Mempreproses gambar menggunakan fungsi yang telah didefinisikan
img_array = preprocess_image(img_path)

# Memprediksi dengan model yang telah dilatih
prediction = model.predict(img_array)

# Pemetaan prediksi numerik ke label kelas
class_labels = {0: "Arborio", 1: "Basmati", 2: "Ipsala", 3: "Jasmine", 4: "Karacadag"}

# Mendapatkan kelas yang diprediksi
predicted_class = np.argmax(prediction, axis=1)[0]

# Mencetak prediksi
print(f'Prediksi: {class_labels[predicted_class]}')
```

```python
# Memuat gambar dari Google Drive
img_path = '/content/drive/MyDrive/GC7/image_2.jpg'

# Mempreproses gambar menggunakan fungsi yang telah didefinisikan
img_array = preprocess_image(img_path)

# Memprediksi dengan model yang telah dilatih
prediction = model.predict(img_array)

# Pemetaan prediksi numerik ke label kelas
class_labels = {0: "Arborio", 1: "Basmati", 2: "Ipsala", 3: "Jasmine", 4: "Karacadag"}

# Mendapatkan kelas yang diprediksi
predicted_class = np.argmax(prediction, axis=1)[0]

# Mencetak prediksi
print(f'Prediksi: {class_labels[predicted_class]}')
```

Hasil *inference* menunjukkan bahwa model berhasil memberikan prediksi yang sesuai untuk **image_1**. Selain itu, model juga berhasil memprediksi hasil yang tepat untuk **image_2**, yang menunjukkan kemampuan model untuk menggeneralisasi pada data baru.
